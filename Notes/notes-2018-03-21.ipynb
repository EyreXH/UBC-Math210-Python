## Euler's method 

Consider a first order differential equation of the form $y' = f(t,y)$. Euler's method is an iterative algorithm for approximating a solution of a first order differential equation with initial value $y(t_0) = y_0$. The formula for Euler's method with time step $h$ is

$$
y_{n+1} = y_n + f(t_n,y_n) h
$$

Each value $y_n$ is an approximation of the true value $y(t_n)$ where $t_n = t_0 +nh$.

Let's write a function which implies implements Euler's method.

```python
def euler_method(f,t_interval,y0,h):
    '''Approximate a solution of y'=f(t,y) by Euler's method.
    
    y[n+1] = y[n] + f(t[n],y[n])*h
    
    parameters
    -------------
    f : function of 2 variables, the right hand side of the ODE y'=f(t,y)
    t : list of length 2 [t0,tf] defining the interval or intergration
    y0 : initial value y(t0) = y0
    h : time step in Euler's method
    
    Returns
    -------------
    A 2D NumPy array where column at index 0 are t values from t0 to tf (excusively) by step h
    and column at index 1 contains y values of a approximation of the solution of y' = f(t,y) 
    with initial condition y(t_0) = y_0. 
    '''
    t = np.arange(t_interval[0],t_interval[1],h)
    N = len(t)
    y = np.zeros(N)
    y[0] = y0
    for n in range(0,N-1):
        y[n+1] = y[n] + f(t[n],y[n])*h
    return np.column_stack([t,y])
```

```python
def f(t,y):
    return y

solution = euler_method(f,[0,1],1,0.1)
plt.plot(solution[:,0],solution[:,1])
```

```python
def f(t,y):
    return np.sin(y)

solution = euler_method(f,[0,5],1,0.001)
plt.plot(solution[:,0],solution[:,1])
```

```python
def f(t,y):
    return np.sin(y)**2 + 2*t**2 - 10*y**2 

solution = euler_method(f,[0,5],1,0.001)
plt.plot(solution[:,0],solution[:,1])
```
