## 1. Plotting with Matplotlib

We've already done some basic plotting with `pyplot` commands. Let's do another example: $y = \frac{1 + x^2}{1 + x + x^3}$ for $x \in [0,2]$

```python
x = np.linspace(0,2,100)
y = (1 + x**2)/(1 + x + x**3)
plt.plot(x,y,'r+')
plt.show()
```

To create a parametrix plot, we create an array of $t$ values nd then arrays of $x$ and $y$.

Let's plot the Trefoil know in 2D:

$$
x(t) = \sin t + 2 \sin 2t
\\
y(t) = \cos t - 2 \cos 2t
$$

for $t \in [0,2 \pi]$ 

```python
t = np.linspace(0,2*np.pi,100)
x = np.sin(t) + 2* np.sin(2*t)
y = np.cos(t) - 2* np.cos(2*t)
plt.plot(x,y)
plt.title('2D Trefoil Knot')
plt.axis('equal')
plt.axis('off')
plt.savefig('trefoil.png',dpi = 200, format = "png")
plt.show()
```

```python
t = np.linspace(0,2*np.pi,100)
x = 16 * np.sin(t)**3
y = 13* np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)
plt.plot(x,y,'r')
plt.title('To Sherry')
plt.axis('equal')
plt.axis('off')
plt.savefig('heart.png',dpi = 200, format = "png")
plt.show()
```
