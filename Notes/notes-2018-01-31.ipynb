### Matplotlib

Let's plot these arrays using matplotlib!

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Let's plot $y = e^(-0.1x^2)cos(2\pi x)$ for $x \in [-10,10]$.

```python
x = np.linspace(-10,10,500)
y = np.exp(0.1*-x**2) * np.cos(2*np.pi*x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-10,10,200)
y = np.exp(0.1*-x**2) * np.cos(2*np.pi*x)
plt.plot(x,y,'g-.')
plt.show()

x = np.linspace(-10,10,200)
y = np.exp(0.1*-x**2) * np.cos(2*np.pi*x)
plt.plot(x,y,'r-.')
plt.show()
```

### SciPy

SciPy has many modules and subpackages for scientific computing such as ODE solvers, optimization, statistics, etc. Let's import the subpackages `special` and plot the plot the error function.
