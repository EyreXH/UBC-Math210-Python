```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Elections Canada Data

```python
data = pd.read_csv('http://www.math.ubc.ca/~pwalls/data/elections.csv')
```

```python
data.info()
```

```python
data.head()
```

Who has run the most time and lost every time?

```python
data.groupby(['Last Name','First Name']).size().sort_values(ascending=False)
```

```python
data[(data['Last Name'] == 'Turmel') & (data['First Name'] == 'John C.')]
```

```python
data['Party'].unique()
```

How many seats did the Liberal win in each election?

```python
data[data['Party'] == 'Liberal'].groupby('Parliament')['Elected'].sum().plot(kind='bar',figsize=(12,5))
```

```python
data[data['Party'] == 'Progressive Conservative'].groupby('Parliament')['Elected'].sum().plot(kind='bar',figsize=(12,5))
```

Which party won each election?

```python
data.groupby(['Parliament','Party'])['Elected'].sum().reset_index()
```

How many different parties won a seat in each election?

```python
elected = data[data['Elected'] ==1]
```

```python
elected.groupby('Parliament')['Party'].nunique().plot(kind='bar',figsize =(12,5))
```
