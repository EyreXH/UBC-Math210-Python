**Question 1**

```python
def spirograph(r,m,a,N):
    '''Plot the parameteric curve:
    
    x = r*(m-1) * np.cos(t/m) + a * np.cos((1 - 1/m)*t)
    y = r*(m-1) * np.sin(t/m) - a * np.sin((1 - 1/m)*t) 
    
    for t in [0,2*pi*m] using N equally spaced points.
    
    Input:
        r,m,a : positive integers, parameters in the parametric equations
        N : positive integer, the number of points used in the plot
        
    Output:
        Plot the parametric equation and return a NumPy array of size N by 2
        where the x values of the plot are the first column and y values are
        the second.
    '''
    t = np.linspace(0,2*np.pi*m,N)
    x = r*(m-1) * np.cos(t/m) + a * np.cos((1 - 1/m)*t)
    y = r*(m-1) * np.sin(t/m) - a * np.sin((1 - 1/m)*t) 
    plt.plot(x,y)
    plt.title('Sprigraph for r={}, m={} and a={}'.format(r,m,a))
    plt.axis('equal'), plt.axis('off'), plt.show()
    M = np.column_stack([x,y]) # Or use np.vstack([x,y]).T
    return M
```

```python
spirograph(10,22,40,1000);
```

**Question 2**

```python
def log_integral(c,N):
    '''Compute the right Riemann sum for the integral of ln(x)/(1 + ln(x)) over the interval [1,c].
    
    Input:
        c : number, the upper limit of integration
        N : integer, the number of subintervals used to compute the Riemann sum
        
    Output:
        The value of the Riemann sum using right endpoints of N subintervals of equal length.
    '''
    dx = (c - 1)/N
    x = np.linspace(1,c,N+1)
    values = np.log(x) / (1 + np.log(x))**2 * dx
    return values.sum()
```

**Question 3**

```python
def cos_product(x,N):
    "Compute partial product \prod_{k=1}^N (1 - 4*x**2/(np.pi**2 * (2*k - 1))**2"
    ks = np.arange(1,N+1)
    values = 1 - 4*x**2 / np.pi**2 / (2*ks - 1)**2
    return np.prod(values)
```

**Question 4**

```python
def a_product(N):
    '''Compute the product a1*a2*...*aN/(2^N) where a1,...,aN are terms the recursive sequence:
    
    a_{k+1} = \sqrt{2 + a_k} , a_1 = sqrt{2}
    
    Input:
        N : positive integer
        
    Output:
        Return the product a1*a2*...*aN/(2^N) 
    '''
    ak = np.zeros(N)
    ak[0] = np.sqrt(2)
    for k in range(1,N):
        ak[k] = np.sqrt(2 + ak[k-1])
    return np.prod(ak) / 2**N
```

**Question 5**

```python
def sum_sine_plot(N,T,m):
    '''Plot the partial sawtooth wave function:
    
    y(t) = 1/2 + 1/pi * sum_{k=1}^{N} sin(2*pi*k*t) / k
    
    for t in [0,2*pi*T].
    
    Input:
        N : positive integer
        T : number
        m : positive integer
    
    Output:
        Plot the function:
        
        y(t) = 1/2 + 1/pi * sum_{k=1}^{N} sin(2*pi*k*t) / k
        
        for t in [0,2*pi*T] and return an m by 2 NumPy array
        where the first column is the t values and the second
        column is the y values.
    '''
    t = np.linspace(0,2*np.pi*T,m)
    terms = np.sum([np.sin(2*np.pi*k*t) / k for k in range(1,N+1)],axis=0)
    y = 1/2 - 1/np.pi * terms
    plt.plot(t,y), plt.xlim([0,2*np.pi*T]), plt.show()
    M = np.column_stack([t,y])
    return M
```

```python
sum_sine_plot(5,0.5,1000);
```
